```python
sse=[]
krng=range(1,10)
for i in krng:
    km=KMeans(n_clusters=i)
    km.fit(df[['Age','Income($)']])
    sse.append(km.inertia_)
sse
#calculate the sum of squared error for 1-10 k values
```

```python
plt.xlabel('k')
plt.ylabel('sum of squeared errror')
plt.plot(krng,sse)
#Eleblow plot for identifing the best k
```
