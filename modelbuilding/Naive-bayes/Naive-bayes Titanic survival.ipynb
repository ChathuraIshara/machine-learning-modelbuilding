```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

```python
df=pd.read_csv('titanic.csv')
df
```

```python
df=df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Fare','Cabin','Embarked'],axis='columns')
```

```python
dummies=pd.get_dummies(df.Sex)
dummies
```

```python
df=pd.concat([df,dummies.female,dummies.male],axis='columns')
df
```

```python
df=df.drop(['Sex'],axis='columns')
df
```

```python
x=df.drop(['Survived'],axis='columns')
x['Age']=x['Age'].fillna(x.Age.median())
#first the na values should be filled with median.other wise our model is not working
```

```python
res=x.Age.fillna(x.Age.median())
res
```

```python
y=df.Survived
y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
model=GaussianNB()
model.fit(xtrain,ytrain)
#GaussianNB model is used for mostly in normal distributions
```

```python
model.score(xtest,ytest)
```
