```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
iris=load_iris()
```

```python
dir(iris)
```

```python
iris.data[0]
```

```python
iris.feature_names
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df
```

```python
df['target']=iris.target
df
```

```python
df['flower']=df['target'].apply(lambda x:iris.target_names[x])
```

```python
df
```

```python
df[df.target==2]
```

```python
df0=df[df.target==0]
df1=df[df.target==1]
df2=df[df.target==2]
```

```python
plt.scatter(x=df0['sepal length (cm)'],y=df0['sepal width (cm)'],color='blue',marker='+')
plt.scatter(x=df1['sepal length (cm)'],y=df1['sepal width (cm)'],color='green',marker='*')
#plt.scatter(x=df2['sepal length (cm)'],y=df2['sepal width (cm)'],color='red',marker='+')

```

```python
from sklearn.svm import SVC
```

```python
x=df.drop(['target','flower'],axis='columns')
x
```

```python
y=df.target
y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
model=SVC(kernel='linear')
model.fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
model.predict([[2.7,3.2,1.3,0.2]])
```
