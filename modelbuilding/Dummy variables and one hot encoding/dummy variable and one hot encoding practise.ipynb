```python
import pandas as pd
```

```python
df=pd.read_csv('homeprices.csv')
df
```

```python
dummies=pd.get_dummies(df.town)
dummies
#generate dummy variables for town nominal variable
```

```python
merged=pd.concat([df,dummies],axis='columns')
merged
#concat df and dummies variable dataframes
```

```python
final=merged.drop(['town','west windsor'],axis='columns')
final
#remove west widsor for avoiding dummy variable trap
#remove town nominal variable
```

```python
from sklearn import linear_model
```

```python
reg=linear_model.LinearRegression()
x=final.drop(['price'],axis='columns')
x
#drop the price and cretaed twod array for training
```

```python
reg.fit(x,df.price)
```

```python
reg.predict([[2800,0,1]])
#predict the price for area 2800 in robinsville
```

```python
reg.predict([[3400,0,0]])
#predic the price for area 3400 in west wisdon
```

```python
reg.score(x,df.price)
#accuracy of our trained model
```
